```python
import numpy as np
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/df_sample.csv')
df.head()
```

```python
# z-표준화

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['중간_z'] = scaler.fit_transform(df[['중간']])
df.head()
```

```python
mid_mean = np.mean(df['중간'])
mid_std = np.std(df['중간'])

df['중간_z2'] = (df['중간']-mid_mean)/mid_std
df.head()
```

```python
# min-max 정규

from sklearn.preprocessing import MinMaxScaler

scaler2 = MinMaxScaler()
df['기말_mm'] = scaler2.fit_transform(df[['기말']])
df.head()
```

```python
last_min = np.min(df['기말'])
last_max = np.max(df['기말'])

df['기말_mm2'] = (df['기말']-last_min)/(last_max-last_min)
df.head()
```
